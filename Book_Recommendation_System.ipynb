### Popularity Based Recommendation System

1.In Popularity Based Recommendation System I will be displaying the top 50 high average rating books.

2.I will be Considering  only the books which are rated by atleast 250 persons because books voted by lesser number of people does not affect the average rating at all.

3.For achieving this I will be Merging the ratings and books dataframe based on the ISBN Column.

```python
ratings_with_name=ratings_df.merge(books_df,on='ISBN') # merging the ratings and books dataframe
```

```python
# Grouping the ratings_with_name dataframe by the title of the books and the number of ratings
grouped_ratings = ratings_with_name.groupby(['Book-Title']).size().reset_index(name='Number-of-Ratings')

# Displaying the grouped ratings
grouped_ratings.head()
```

```python
grouped_ratings
```

```python
average_rating=ratings_with_name.groupby(['Book-Title'])["Book-Rating"].mean().reset_index(name="Average-Ratings") # Grouping the ratings_with_name dataframe by the title of the books and the average ratings
```

```python
average_rating.head()
```

```python
popularity_df=grouped_ratings.merge(average_rating,on="Book-Title") # Merging the grouped_ratings and average_ratings dataframe
```

```python
# Selecting rows with number of ratings greater than 250
filtered_df=popularity_df[popularity_df['Number-of-Ratings']>=250]

# Sorting the filtered DataFrame in descending order of average rating
sorted_df=filtered_df.sort_values(by='Average-Ratings', ascending=False)

# Displaying the sorted DataFrame
sorted_df
```

```python
sorted_df.merge(books_df,on="Book-Title") # Merging the sorted_df and books_df dataframe
```

```python
sorted_df=sorted_df.merge(books_df,on="Book-Title").drop_duplicates(subset=["Book-Title"]) # Merging the sorted_df and books_df dataframe and dropping the duplicate rows
```

#### Collaborative Filtering Based Book Recommendation System

 1.In Colloborative Filtering based Book recommmendation system, we will be considering all the users and all the books and we will be recommending books based on the ratings given by the users to the books.

2.We will be consider only the users who have given more than 200 ratings and the books which have more than 100 ratings.




```python
x=ratings_with_name.groupby('User-ID').count()["Book-Rating"]>200 # Grouping the ratings_with_named dataframe by the User-Id and counting the number of ratings given by each user and taking the books whc have  greater than 200 ratings
y=x[x].index.tolist() # Converting the x into a list
```

```python
y[:5]
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(y)] # Selecting the rows with the User-Id in y
```

```python
filtered_rating.shape
```

```python
top_books=filtered_rating.groupby('Book-Title').count()["Book-Rating"].sort_values(ascending=False)>=50 # Grouping the filtered_rating dataframe by the Book-Title and counting the number of ratings given to each book and taking the books which have greater than 50 ratings.
```

```python
top_books=top_books[top_books].index.tolist() # Converting the top_books into a list
```

```python
final_ratings=filtered_rating[filtered_rating["Book-Title"].isin(top_books)]
```

```python
final_ratings
```

```python
pivot_t=final_ratings.pivot_table(index="Book-Title",columns='User-ID',values='Book-Rating') # Creating a pivot table with index as Book-Title, columns as User-Id and values as Book-Rating
```

```python
pivot_t.head()
```

```python
pivot_t.shape
```

```python
pivot_t.fillna(0,inplace=True)# Fill the NaN values with 0.
```

```python
from sklearn.metrics.pairwise import cosine_similarity #Importiing the cosine_similarity function from sklearn.metrics.pairwise.
```

```python
similarity_score=cosine_similarity(pivot_t)# calaculating the cosine similarity of the pivot_t dataframe.
```

```python
similarity_score[:5]
```

```python
# Creating the recommend function which takes the book names as input and returns the top 10 similar books.
def recommend(book_name):
    if book_name not in pivot_t.index: # Checking if the book is in the list
        return "Book is not in the list"
    else:
        book_id=pivot_t.index.get_loc(book_name) # Getting the index of the book
        book_list=list(enumerate(similarity_score[book_id])) # Getting the list of books with their similarity score
        book_list=sorted(book_list,key=lambda x:x[1],reverse=True) # Sorting the book_list in descending order of similarity score
        book_list=book_list[1:11] # Selecting the top 10 books
        recommended_books=[i[0] for i in book_list] # Getting the index of the top 10 books
        return pivot_t.index[recommended_books]
```

```python
recommend("The Queen of the Damned (Vampire Chronicles (Paperback))")
```
